```python
# Cell 1 — installs
!pip install pandas numpy scikit-learn lightgbm matplotlib seaborn requests aiohttp understat==0.1.1 joblib

# Cell 2 — imports
import os, io, requests, math, datetime
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import log_loss, accuracy_score, classification_report
import lightgbm as lgb
import matplotlib.pyplot as plt
import joblib

# Cell 3 — helper: download historical CSVs from football-data (example: 2015-2025)
# We will programmatically download a few seasons. If blocked, upload CSVs manually.
base = "https://www.football-data.co.uk/mmz4281"
seasons = ["1516","1617","1718","1819","1920","2021","2122","2223","2324","2425"]  # adapt if needed
frames = []
for s in seasons:
    url = f"{base}/{s}/E0.csv"
    try:
        r = requests.get(url, timeout=15)
        r.raise_for_status()
        df = pd.read_csv(io.StringIO(r.text))
        df['Season'] = s
        frames.append(df)
        print("Downloaded", s)
    except Exception as e:
        print("skip", s, e)
all_df = pd.concat(frames, ignore_index=True)
print("combined rows:", len(all_df))

# Cell 4 — basic cleaning & canonical columns
# football-data CSVs use columns: Date, HomeTeam, AwayTeam, FTHG, FTAG, FTR (H/D/A), and odds columns
df = all_df.rename(columns={
    'Date':'date','HomeTeam':'home','AwayTeam':'away','FTHG':'home_goals','FTAG':'away_goals','FTR':'result'
})
df = df[['date','Season','home','away','home_goals','away_goals','result'] + [c for c in df.columns if 'Bb' in c][:6]]
df['date'] = pd.to_datetime(df['date'], dayfirst=True, errors='coerce')
df = df.dropna(subset=['date'])
df = df.sort_values('date').reset_index(drop=True)
print("clean df:", df.shape)
df.head()

# Cell 5 — create target & some rolling features
def encode_result(r):
    if r=='H': return 0  # home win
    if r=='D': return 1  # draw
    return 2              # away win

df['target'] = df['result'].map(encode_result)

# rolling team stats: recent 5-match points, goals for/against average
def team_rolling_features(df, window=5):
    teams = pd.unique(df[['home','away']].values.ravel('K'))
    # initialize containers
    feats = []
    team_stats = {t: {'played':0,'points':0,'gf':0,'ga':0,'recent':[]} for t in teams}
    for idx,row in df.iterrows():
        h,a = row['home'], row['away']
        # compute features based on current team_stats
        h_st = team_stats[h]
        a_st = team_stats[a]
        feats.append({
            'index': idx,
            'home_points_5': sum(h_st['recent']) if h_st['recent'] else 0,
            'away_points_5': sum(a_st['recent']) if a_st['recent'] else 0,
            'home_gf_avg': (h_st['gf']/h_st['played']) if h_st['played']>0 else 0,
            'home_ga_avg': (h_st['ga']/h_st['played']) if h_st['played']>0 else 0,
            'away_gf_avg': (a_st['gf']/a_st['played']) if a_st['played']>0 else 0,
            'away_ga_avg': (a_st['ga']/a_st['played']) if a_st['played']>0 else 0,
        })
        # update stats with this match result for future matches
        hg, ag = row['home_goals'], row['away_goals']
        # points
        if hg>ag:
            hp, ap = 3,0
        elif hg==ag:
            hp, ap = 1,1
        else:
            hp, ap = 0,3
        for team,pts,gf,ga in [(h,hp,hg,ag),(a,ap,ag,hg)]:
            st = team_stats[team]
            st['played'] += 1
            st['points'] += pts
            st['gf'] += gf
            st['ga'] += ga
            st['recent'].append(pts)
            if len(st['recent'])>window:
                st['recent'].pop(0)
    feats_df = pd.DataFrame(feats).set_index('index')
    return pd.concat([df.reset_index(drop=True), feats_df.reset_index(drop=True)], axis=1)

df = team_rolling_features(df, window=5)
df = df.dropna(subset=['home_gf_avg'])  # drop early
print("after features:", df.shape)
df.head()

# Cell 6 — optional: fetch Understat xG features for teams (may be slow/rate-limited).
# If understat fetch works, we'll add season-level xG metrics for teams to improve predictions.
try:
    from understat import Understat
    import aiohttp, asyncio
    async def fetch_season_xg(season='2024'):
        async with aiohttp.ClientSession() as session:
            u = Understat(session)
            teams = await u.get_league_teams('EPL', season)
            return teams
    loop = asyncio.get_event_loop()
    recent_season = str(int(df['Season'].iloc[-1][:2]) + 2000)
    teams_xg = loop.run_until_complete(fetch_season_xg('2024'))
    xg_map = {t['title']: float(t.get('xG','0')) for t in teams_xg}
    df['home_xg_season'] = df['home'].map(xg_map).fillna(0)
    df['away_xg_season'] = df['away'].map(xg_map).fillna(0)
    print("Understat features added")
except Exception as e:
    print("Understat fetch failed or not available:", e)
    df['home_xg_season'] = 0
    df['away_xg_season'] = 0

# Cell 7 — prepare training data
feature_cols = [
    'home_points_5','away_points_5',
    'home_gf_avg','home_ga_avg','away_gf_avg','away_ga_avg',
    'home_xg_season','away_xg_season'
]
X = df[feature_cols]
y = df['target']
print("X shape", X.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.15, random_state=42)

# Cell 8 — lightgbm training
train_data = lgb.Dataset(X_train, label=y_train)
valid_data = lgb.Dataset(X_test, label=y_test, reference=train_data)
params = {
    'objective':'multiclass',
    'num_class':3,
    'metric':'multi_logloss',
    'learning_rate':0.05,
    'num_leaves':31,
    'verbose':-1,
    'seed':42
}
bst = lgb.train(
    params,
    train_data,
    num_boost_round=500,
    valid_sets=[valid_data],
    callbacks=[
        lgb.early_stopping(stopping_rounds=30),
        lgb.log_evaluation(period=50)
    ]
)

# Cell 9 — evaluation
probs = bst.predict(X_test)
y_pred = np.argmax(y_prob, axis=1)
print("Log loss:", log_loss(y_test, y_prob))
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=['home','draw','away']))

# Cell 10 — Predict the opener: Liverpool vs Bournemouth
# Create a feature vector for the specific match using latest stats from df
def build_match_row(team_home, team_away, df_full):
    # use most recent rolling stats for the two teams (last available)
    last = df_full.copy()
    # find last match stats by team
def last_stats(team):
    subset = last[(last['home'] == team) | (last['away'] == team)]
    if subset.empty:
        # No data for this team, return zeros
        return {
            'points_5': 0,
            'gf_avg': 0,
            'ga_avg': 0,
            'xg_season': 0
        }
    r = subset.iloc[-1]
    return {
        'points_5': r['home_points_5'] if r['home'] == team else r['away_points_5'],
        'gf_avg': r['home_gf_avg'] if r['home'] == team else r['away_gf_avg'],
        'ga_avg': r['home_ga_avg'] if r['home'] == team else r['away_ga_avg'],
        'xg_season': r['home_xg_season'] if r['home'] == team else r['away_xg_season']
    }


# Cell 11 — quick plot for reel
plt.figure(figsize=(6,3))
labels = ['Home (Liverpool)','Draw','Away (Bournemouth)']
plt.bar(labels, probs[0])
plt.ylim(0,1)
plt.title("Model prediction: Liverpool vs Bournemouth (home/draw/away)")
plt.ylabel("Probability")
plt.tight_layout()
plt.savefig("prediction_bar.png", dpi=150)
plt.show()

# Cell 12 — save results
pd.DataFrame([probs[0]], columns=['P_home','P_draw','P_away']).to_csv("opener_prediction.csv", index=False)
print("Saved opener_prediction.csv and prediction_bar.png")
```
